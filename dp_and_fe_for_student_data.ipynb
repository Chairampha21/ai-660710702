# จัดการกับข้อมูลในคอลัมน์ ชื่อ ที่มีชื่อเป็น NaN โดยใช้คำสั่ง df.dropna(subset=['ชื่อ'])

* ใช้คำสั่ง df.dropna(subset=['ชื่อ']) เพื่อลบแถวทั้งหมดที่มีค่า NaN (ค่าหาย) ออกจากข้อมูลไฟล์ student.csv
* เนื่องจากคอลัมน์ "ชื่อ" เป็นข้อมูลสำคัญที่ควรมีอยู่ครบ ไม่สามารถนำค่าเฉลี่ยหรือค่าสถิติอื่นมาทดแทนได้เหมือนกับคอลัมน์อายุจึงควรลบแถวนั้นออกไป

```python
df_clean = df.dropna(subset=['ชื่อ'])
```

```python
df_original = df
df = df_clean
df
```

# จัดการกับข้อมูลในคอลัมน์ อายุ ซึ่งเป็นการจัดการค่าผิดปกติ
*   ในคอลัมน์อายุมีอายุ 150 ซึ่งเป็นค่าที่ผิดปกติจึงควรตัดออกเพื่อให้ได้ข้อมูลที่ถูกต้องสมบูรณ์
#โดยทำการกรองค่าผิดปกติเพื่อตัดแถวที่มีค่าอายุเป็น 150 ปี ให้หายไป


```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_no_outliers = df[(df['อายุ'].isna()) | ((df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound))]
```

```python
df = df_no_outliers
df
```

# จัดการกับค่า NaN ในคอลัมน์ อายุ โดยใช้ค่าเฉลี่ยและนำข้อมูลมาเติมค่าลงไปแทนค่าNaN

```python
mean_age = df['อายุ'].mean()
df['อายุ'].fillna(mean_age, inplace=True)
```

```python
df
```

# ทำให้ข้อมูลอายุเป็นค่าจำนวนเต็มเนื่องจากค่าอายุดูไม่ค่อยมีความเหมาะสมกับการดูข้อมูล จึงต้องการทำให้เป็นจำนวนเต็มเพื่อให้ง่ายต่อการอ่านข้อมูล

```python
df['อายุ'] = np.floor(df['อายุ']).astype(int)
df
```

# จากคอลัมน์ เพศ จะเห็นว่าข้อมูลเพศใช้ข้อมูลกระจายหลายรูปแบบมากเกินไป มีทั้ง ชาย หญิง หรือ ช ญ Female F M จึงควรทำข้อมูลให้ข้อมูลเป็นมาตรฐานเดียวกันคือใช้ ชาย และ หญิง

```python
def standardize_gender(gender):
    if gender in ['ชาย', 'ช', 'Male', 'M']:
        return 'ชาย'
    elif gender in ['หญิง', 'ญ', 'Female', 'F']:
        return 'หญิง'
    else:
        return gender

df['เพศ'] = df['เพศ'].apply(standardize_gender)
```

```python
df
```

# จัดการกับค่า  NaN ในคอลัมน์เพศ โดยใช้การเติมค่าด้วยค่าที่พบบ่อยที่สุดลงไปแทนค่า NaN

```python
df['เพศ'].fillna(df['เพศ'].mode()[0], inplace=True)
```

```python
df
```

# จัดการกับคอลัมน์ คณะ	 ทำให้ข้อมูลเป็นมาตรฐานแบบเดียวกัน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์' ,'Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts' , 'Liberal Arts','คณะศิลปศาสตร์']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวฯ', 'วิศวะ','วิศวกรรม','Engineering', 'คณะวิศวกรรมศาสตร์']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name

# ใช้งาน
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

# จัดการกับค่า  NaN ในคอลัมน์คณะ โดยใช้การเติมค่าด้วยค่าที่พบบ่อยที่สุดลงไปแทนค่า NaN

```python
df['คณะ'].fillna(df['คณะ'].mode()[0], inplace=True)
```

```python
df
```

# จัดการกับค่าผิดปกติในคอลัมน์ GPA ที่มีค่าไม่ปกติ คือ -1.00 และ 4.50	เพราะgpa ไม่สามารถ ติดลบ หรือเกิน 4.00 ได้ ทำให้เป็นค่า NaN

```python
df.loc[(df['GPA'] < 0) | (df['GPA'] > 4.00), 'GPA'] = np.nan
```

```python
df
```

# จัดการกับค่า NaN ในคอลัมน์ GPA โดยใช้ค่าเฉลี่ยและนำข้อมูลมาเติมค่าลงไปแทนค่าNaN

```python
mean_gpa = round(df['GPA'].mean(), 2)
df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
df
```

# จัดการกับค่า NaN ในคอลัมน์ ส่วนสูง_cm โดยใช้ค่าเฉลี่ยและนำข้อมูลมาเติมค่าลงไปแทนค่าNaN

```python
mean_height = df['ส่วนสูง_cm'].mean()
df['ส่วนสูง_cm'].fillna(mean_height, inplace=True)
```

```python
df
```

# จัดการกับค่า NaN ในคอลัมน์ น้ำหนัก_kgb โดยใช้ค่าเฉลี่ยและนำข้อมูลมาเติมค่าลงไปแทนค่าNaN

```python
mean_weight = round(df['น้ำหนัก_kg'].mean(), 1)
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)
```

```python
df
```

# จัดการกับค่า NaN ในคอลัมน์ ชั่วโมงเรียน_ต่อสัปดาห์ โดยใช้ค่าเฉลี่ยและนำข้อมูลมาเติมค่าลงไปแทนค่าNaN

```python
mean_time = round(df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean(), 1)
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_time, inplace=True)
```

# การสร้าง Feature จาก GPA

```python
data['ระดับผลการเรียนเฉลี่ยสะสม'] = pd.cut(data['GPA'],
    bins=[-float('inf'), 2.00, 3.20, float('inf')],
    labels=['รอพินิจ', 'ปกติ', 'เรียนเด่น'],
    right=False
)
```

```python
data
```

# การรวม Features ของข้อมูลน้ำหนักและส่วนสูงมาหาค่า BMI และแสดงผลสุขภาพได้นั่นเอง

```python
data['BMI'] = data['น้ำหนัก_kg'] / (data['ส่วนสูง_cm']/100)**2
data['สุขภาพ'] = pd.cut(data['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
```
